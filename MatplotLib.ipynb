```python
import matplotlib.pyplot as plt  #in matplotlib mainly we use pyplot for plotting the graphs
import numpy as np
import matplotlib as mp
```

```python
#Genenrating sine and cos graph using mp
#plt.legend(loc='upper right/left, lower left/right)
#plt.title()
x = np.linspace(0,4*np.pi,50)
cos_y = np.cos(x)
sin_y = np.sin(x)
# print(sin_y)

plt.plot(x,sin_y, label='Sine')
plt.plot(x,cos_y, label='Cosine')
plt.title('Sin_Cos_wave')
plt.legend(loc='lower left')
plt.show()
```

```python
plt.plot(x,sin_y, '*')
plt.plot(x,cos_y, '--')
```

Scatter Plots--is a type of Data visualisation in which individual data are represented as points in a 2-D plane.
The x-coordinate of each point represents one feature of the data, and the y-coordinate represents another feature.
Scatter plots are usually used to visualize the relationship between two varibales and to identify patterns or outliers 
in the data.


```python
x=np.random.randn(500)
y=np.random.randn(500)
plt.figure(figsize=[5,4])
plt.scatter(x,y, c='red', alpha=0.8, s=6)  #aplha-color of the dots in the plot 1-darkest, 0-lightest  #s is the size of the dots

plt.title('Scatter Plot')
plt.show()
```

```python
n=1000
x1 = np.random.randn(n)
y1=np.random.randn(n)

x2 = np.random.randn(n)+2   # +2 is bcz it destinguishes both the features here(red and black)
y2 = np.random.randn(n)+2

plt.figure(figsize=[4,5])
plt.scatter(x1, y1, c='red', s=6, alpha=0.7)  #c=color, s=sizes alpha=size of the the dots or marks in the plot
plt.scatter(x2, y2, c='black',s=4,alpha=1)
plt.title('Scatter plot with multiple groups')
plt.show()
```

```python
#Random Scatter Plot
rng = np.random.RandomState(0) 
#RandomState(0) ensures that the random numbers generated will be the same every time the code is run.
x=rng.randn(100)
y=rng.randn(100)

colors = rng.rand(100)
size = rng.rand(100)*1000
plt.figure(figsize=[7,5])
# plt.xlim=([0,10])
# plt.ylim=([-2,2])
plt.scatter(x,y, c=colors, s=size, alpha=.7, cmap='viridis', edgecolors='red' )  #cmap=color map
plt.title("random scatter plot")
plt.xlabel('x values')
plt.ylabel('y values')

#for showing the density of color bar
plt.colorbar()
plt.show()
```

Histograms and Binnings

```python
data = np.random.randn(100)  #generates 100 random number. Diiferent every time we run the code(because no RandomState())
data
```

```python
data = np.random.randn(100000)
plt.figure(figsize=[4,3])
plt.hist(data, bins=1000, edgecolor='black', alpha=0.8, color='orange')
# Creating a histogram with 1000 bins, black edges, 80% transparency, and orange color
#bins=1000 means the data will be divided into 1000 small intervals, which makes the histogram more detailed.
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
data = np.random.randn(100000)
plt.figure(figsize=[4,3])
plt.hist(data, bins=100, edgecolor='black', alpha=0.8, color='orange', histtype='stepfilled')
#stepfilled it will remove the edge of each bin and adds all to a total removes all the in-between bars
plt.title("Random Histogram")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plotting Multiple histogram
x1 = np.random.randn(100000)
x2 = np.random.randn(100000)+1
x3 = np.random.randn(100000)+3
plt.figure(figsize=[4,3])
plt.hist(x1, bins=50, edgecolor='black', alpha=0.8, color='orange')
plt.hist(x2, bins=50, edgecolor='black', alpha=0.8, color='red')
plt.hist(x3, bins=50, edgecolor='black', alpha=0.8, color='green')


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Multiple Histogram(side view)')
plt.show()
```

2D Histograms- This creates a heatmap-A 2D histogram essentially represents the density of data points and provides a top-down view of the distribution. 
--A 2D histogram is used to visualize the relationship between two numerical variables by dividing the data into bins along both the x-axis and y-axis. Instead of showing a single distribution (like a 1D histogram), it represents the frequency of data points across two dimensions.

```python
x = np.random.randn(100000)
y = np.random.randn(100000)

plt.figure(figsize=[4,3])
plt.hist2d(x,y, bins=50, cmap='Oranges')
plt.colorbar(label='Density')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('2D Histogram(top view)')
plt.show()
```

Pie Charts- Circular Statistical graphic,which is divided into slices to illustrate numerical proportion

```python
car=['BMW','Audi','Mercedes','Mazda','Jaguar']
data=[450, 500, 270, 30, 230]
fig = plt.figure(figsize=[4,3])
plt.pie(data, labels=car)
plt.show()

#data = define the sizes of the slices
#labels = gives name to the slices
```

Basic Operation on Pie chart

```python
#explode()

car=['BMW','Audi','Mercedes','Mazda','Jaguar']
data=[450, 500, 270, 30, 230]
explode=[0,0,0,0,.3]
fig = plt.figure(figsize=[4,3])
plt.pie(data, labels=car, explode=explode)
plt.show()
```

```python
#2 Shadows
car=['BMW','Audi','Mercedes','Mazda','Jaguar']
data=[450, 500, 270, 30, 230]
fig = plt.figure(figsize=[4,3])
plt.pie(data, labels=car, shadow=True)
plt.show()
```

```python
#3 Adding Percentage

car=['BMW','Audi','Mercedes','Mazda','Jaguar']
data=[450, 500, 270, 30, 230]
fig = plt.figure(figsize=[4,3])
plt.pie(data, labels=car, autopct='%1.1f%%')
# autopct='%1.1f%%' - Displays the percentage with one decimal place inside each slice
plt.show()
```

```python
# Adding the Wedget
car=['BMW','Audi','Mercedes','Mazda','Jaguar']
data=[450, 500, 270, 30, 230]
fig = plt.figure(figsize=[4,3])
plt.pie(data, labels=car, autopct='%1.1f%%', wedgeprops={'edgecolor':'black','linewidth':0.01,'antialiased':False})
#wedgeprops is a parameter in plt.pie() used to customize the appearance of the slices (wedges) in a pie chart.
# antialiased - Disable anti-aliasing for the edges
#linewidth - Set the thickness of the slice borders
plt.show()
```

Multiple Subplots()

Multiple Subplots - Multiple subplots mean displaying more than one graph (or plot) in a single figure. This helps compare different data visualizations side by side instead of opening multiple windows.

```python
x=np.linspace(0,10, 100)
y1,y2,y3,y4 = np.sin(x), np.cos(x), np.tan(x), -np.sin(x)

fig, axes = plt.subplots(2,2, figsize=[7,6])

axes[0,0].plot(x,y1, color='red')
axes[0,0].set_title('sine wave')

axes[0,1].plot(x,y2, color='blue')
axes[0,1].set_title('cosine wave')

axes[1,0].plot(x,y3, color='orange')
axes[1,0].set_title('tan wave')

data = np.random.rand(1000)
axes[1,1].hist(data, bins=100)
axes[1,1].set_title('histogram')

plt.tight_layout()
plt.show()

#when the x-axis and y-axis are same then it will plot histogram else scatterplot will be plotted.

```

```python
#using loops

fig, axes = plt.subplots(2,2, figsize=[9,5])
plt.subplots_adjust(hspace=.4, wspace=.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5, 0.5, str((2,3,i)), fontsize=15, ha = 'center')
```

```python
#with loops example 2
x=np.linspace(0,10,50)
y=np.sin(x)
    
fig, axes = plt.subplots(1,3, figsize=(10,4))
axes[0].plot(x,y)
axes[1].plot(x,y**2)
axes[2].plot(x,y**3)
    # The axes object is not a 2D array because plt.subplots(1,3) creates a 1-row, 3-column layout, 
    # resulting in a 1D array. You should use axes[i] instead of axes[row, column].
    
axes[0].set_title('sine wave')
axes[1].set_title('sine wave-squared')
axes[2].set_title('sine wave-cubed')
    
plt.tight_layout()
plt.show()
    
    #NOTE: The error 'numpy.ndarray' object has no attribute 'subplot' happens because
    # axes is a NumPy array (when multiple subplots are created),and you're trying to call .subplot() on it, 
    # which does not exist.
```
